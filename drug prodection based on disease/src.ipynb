Visual inspection of each column with a histogram

Separating the target column from the rest of the columns with which we want to make predictions.

```python
x_ = data.iloc[:,0:131].values
y_ = data["prognosis"].values
```

```python
x_[5]
```

```python
y_[5]
```

Data division into two parts: validation and train

```python
x_train , x_validation , y_train , y_validation =train_test_split(x_,y_,test_size=0.20,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
print(x_validation.shape,y_validation.shape)
```

# Modeling

**KNeighborsClassifier**

Checking which value of **k** to choose

```python
kn = 10
accuracy = np.zeros((kn-1))

for number in range(1,kn):
    neighbors = KNeighborsClassifier(n_neighbors=number)
    neighbors.fit(x_train,y_train)
    yhat_knn =neighbors.predict(x_validation)
    accuracy[number-1]=metrics.accuracy_score(y_validation,yhat_knn)
accuracy
```

We choose k=3 randomly

```python
neighbor = KNeighborsClassifier(n_neighbors=3)
neighbor.fit(x_train,y_train)
```

validation for **k=3**

```python
yhat_k3=neighbor.predict(x_validation)
acc_k3=metrics.accuracy_score(y_validation, yhat_k3)
acc_k3
```

**GradientBoostingClassifier**

We try to make a better model by choosing the right parameters for n_estimators and max_depth

```python
gradientB =GradientBoostingClassifier(n_estimators=100, max_depth =3)
gradientB.fit(x_train, y_train)
```

prediction and vlidation for GradientBoostingClassifier

```python
yhat_gb=gradientB.predict(x_validation)
yhat_gb
```

```python
acc_GBC=metrics.accuracy_score(y_validation, yhat_gb)
acc_GBC
```

Get report for GradientBoostingClassifier **(precision,recall,f1-score,support)**


```python
from sklearn.metrics import classification_report
print (classification_report(y_validation, yhat_gb))
```

**SVM**

```python
sVm= svm.SVC(kernel='linear')
```

```python
sVm.fit(x_train, y_train)
```

prediction and validation svm

```python
yhat_svm = sVm.predict(x_validation)
yhat_svm [0:5]
```

```python
acc_sVm=metrics.accuracy_score(y_validation,yhat_svm)
acc_sVm
```

Get report of model **(precision,recall,f1-score,support)**

```python
print(classification_report(y_validation, yhat_svm))
```

**RandomForestClassifier**

```python
forestclass = RandomForestClassifier()
forestclass
```

```python
forestclass.fit(x_train,y_train)
```

prediction and validation for RandomForestClassifier

```python
yhat_f = forestclass.predict(x_validation)
yhat_f[:5]
```

```python
acc_forest=metrics.accuracy_score(y_validation,yhat_f)
acc_forest
```

Get report of model **(precision,recall,f1-score,support)**

```python
print(classification_report(y_validation, yhat_f))
```
